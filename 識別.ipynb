```python
from PIL import Image, ImageDraw

# 打开地图图片
img = Image.open('d:/study/研一上/CHC5904/demo/map.webp')

# 创建绘图对象
draw = ImageDraw.Draw(img)

# 标记点的位置
x, y = (2185,3490)

# 画一个红色圆点
radius = 30
draw.ellipse([x-radius, y-radius, x+radius, y+radius], fill='red', outline='red')

# 显示图片
img.show()

# 如果需要保存
# img.save('d:/study/研一上/CHC5904/demo/map_marked.png')
```
